```python
def generate_mult_dataset(ruta_dir,file_name, num_G,num_SubG, num_dataset, test_size):
    df = pd.read_csv(ruta_dir+"/"+file_name,header=None,sep='\t')
    df.columns= ['grafo','subgrafo']
    df['grafo'] = [str(df['grafo'][i]).zfill(num_G) for i in range(len(df['grafo']))]
    df['subgrafo'] = [str(df['subgrafo'][i]).zfill(num_SubG) for i in range(len(df['subgrafo']))]
    for i in range(num_dataset):
        X_train, X_test, y_train, y_test = train_test_split(df['grafo'], df['subgrafo'], test_size=1-test_size)
        df_train = pd.concat([X_train, y_train], axis=1)
        df_test = pd.concat([X_test, y_test], axis=1)
        ruta = ruta_dir + "/dataset"+str(i);
        os.makedirs(ruta, exist_ok=True)
        r_train = ruta+"/train"+str(i)+".csv"
        df_train.to_csv(r_train,index=None,header=False,sep='\t',line_terminator='\r\n')
        r_test = ruta+"/test"+str(i)+".csv"
        df_test.to_csv(r_test,index=None,header=False,sep='\t',line_terminator='\r\n')
        
        row = str(len(df_train))+" "+str(num_G)+" "+str(num_SubG)+"\n"
        
        f = open(r_train,'r')
        temp = f.read()
        f.close()

        f = open(r_train, 'w')
        f.write(row)

        f.write(temp)
        f.close()
        
        
        row = str(len(df_test))+" "+str(num_G)+" "+str(num_SubG)+"\n"
        
        f = open(r_test,'r')
        temp = f.read()
        f.close()

        f = open(r_test, 'w')
        f.write(row)

        f.write(temp)
        f.close()
        
        
        
        del df_train
        del df_test
    del df
    
        
        
    
    
```

```python
generate_mult_dataset('./DATASET-10_3','data-10-3.csv',10,3,10,0.7)
```

```python
generate_mult_dataset('./DATASET-10_6','data-10-6.csv',10,6,10,0.7)
```
